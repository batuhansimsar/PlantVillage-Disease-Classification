```
# -*- coding: utf-8 -*-
"""
ConvNeXt-Tiny ile PlantVillage Dataset'inde Elma SÄ±nÄ±flandÄ±rmasÄ±
4 SÄ±nÄ±flÄ± Multi-Class Classification (Apple Diseases)

GELÄ°ÅTÄ°RÄ°LMÄ°Å SÃœRÃœM v2
- Two-Phase Transfer Learning (Phase-1: Only Classifier, Phase-2: Fine-Tuning son bloklar)
- AdamW optimizer + Cosine Annealing LR (Phase-2)
- Label Smoothing ile CrossEntropyLoss
- RandomResizedCrop + geliÅŸmiÅŸ augmentations
- Mixed Precision (AMP) ile hÄ±zlÄ± eÄŸitim (CUDA varsa)
- Google Colab iÃ§in optimize (CUDA GPU desteÄŸi)
"""

import os
import time
import glob
from collections import defaultdict

# Colab ortamÄ± mÄ± kontrol et ve Drive mount
IN_COLAB = False
try:
    from google.colab import drive
    IN_COLAB = True
    drive.mount('/content/drive', force_remount=False)
    print("âœ… Google Drive baÄŸlandÄ±")
except Exception:
    print("âš ï¸  Google Colab ortamÄ± deÄŸil, Drive mount atlandÄ±")

# PyTorch import'larÄ± - Runtime restart kontrolÃ¼ ile
try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from torch.utils.data import DataLoader, random_split, Dataset
    from torchvision import transforms, models
    from torchvision.models import ConvNeXt_Tiny_Weights
    print("âœ… PyTorch ve Torchvision baÅŸarÄ±yla yÃ¼klendi")
except RuntimeError as e:
    if "TORCH_LIBRARY" in str(e):
        print("âŒ HATA: PyTorch Ã§akÄ±ÅŸmasÄ± tespit edildi!")
        if IN_COLAB:
            print("   Runtime otomatik olarak yeniden baÅŸlatÄ±lÄ±yor...")
            print("   LÃ¼tfen bu hÃ¼creyi tekrar Ã§alÄ±ÅŸtÄ±rÄ±n!")
            os.kill(os.getpid(), 9)  # Runtime'Ä± restart et
        else:
            print("   LÃ¼tfen Colab'da Runtime > Restart Runtime yapÄ±n ve tekrar Ã§alÄ±ÅŸtÄ±rÄ±n.")
            raise
    else:
        raise
except Exception as e:
    print(f"âŒ Import hatasÄ±: {e}")
    raise

from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# --- 1. AYARLAR ---

# Transfer Learning: True = HÄ±zlÄ± eÄŸitim (Pretrained), False = SÄ±fÄ±rdan
USE_TRANSFER_LEARNING = False

# PlantVillage dataset yolu - Colab iÃ§in
# Ã–rn: "/content/drive/MyDrive/PlantVillage/color"
DATA_DIR = "/content/drive/MyDrive/color"  # BURAYI KENDÄ° YOLUNA GÃ–RE GÃœNCELLE

# EÄŸitim parametreleri
NUM_EPOCHS_PHASE1 = 40        # Sadece classifier eÄŸitimi
NUM_EPOCHS_PHASE2 = 10        # Fine-tuning (son bloklar)
BATCH_SIZE = 32               # Colab CUDA iÃ§in optimize
BASE_LR_PHASE1 = 1e-4         # Phase-1 LR (sadece classifier)
BASE_LR_PHASE2 = 5e-6         # Phase-2 LR (fine-tuning)
WEIGHT_DECAY = 5e-4
USE_MIXED_PRECISION = True    # Colab CUDA iÃ§in hÄ±z ve bellek optimizasyonu

# Apple sÄ±nÄ±flarÄ±
APPLE_CLASSES = [
    'Apple___Apple_scab',
    'Apple___Black_rot',
    'Apple___Cedar_apple_rust',
    'Apple___healthy'
]


# --- 2. DATASET SINIFI ---

class AppleDataset(Dataset):
    """Elma verilerini 4 sÄ±nÄ±flÄ± multi-class olarak yÃ¼kleyen custom dataset"""
    def __init__(self, data_dir, transform=None):
        self.data_dir = data_dir
        self.transform = transform
        self.samples = []
        self.class_to_idx = {
            'Apple___Apple_scab': 0,
            'Apple___Black_rot': 1,
            'Apple___Cedar_apple_rust': 2,
            'Apple___healthy': 3
        }
        self.idx_to_class = {v: k for k, v in self.class_to_idx.items()}

        found_any = False
        print("\n--- Veri KlasÃ¶rleri TaranÄ±yor ---")
        for folder_name in APPLE_CLASSES:
            folder_path = os.path.join(data_dir, folder_name)
            if os.path.exists(folder_path):
                # FarklÄ± uzantÄ±larÄ± dene
                images = (
                    glob.glob(os.path.join(folder_path, '*.JPG')) +
                    glob.glob(os.path.join(folder_path, '*.jpg')) +
                    glob.glob(os.path.join(folder_path, '*.jpeg')) +
                    glob.glob(os.path.join(folder_path, '*.png'))
                )

                if len(images) == 0:
                    try:
                        all_files = os.listdir(folder_path)
                        print(f"âš ï¸  {folder_name}: KlasÃ¶r var ama resim bulunamadÄ±")
                        print(f"   KlasÃ¶rdeki dosya sayÄ±sÄ±: {len(all_files)}")
                        if len(all_files) > 0:
                            print(f"   Ä°lk 5 dosya: {all_files[:5]}")
                    except Exception:
                        pass
                else:
                    label = self.class_to_idx[folder_name]
                    image_count = 0
                    for img_path in images:
                        self.samples.append((img_path, label))
                        image_count += 1
                    print(f"âœ… {folder_name}: {image_count} resim bulundu")
                    found_any = True
            else:
                print(f"âŒ KlasÃ¶r bulunamadÄ±: {folder_path}")

        print(f"\nğŸ“Š Toplam Ã¶rnek sayÄ±sÄ±: {len(self.samples)}")
        if len(self.samples) == 0:
            print("\nâŒ HATA: HiÃ§ resim bulunamadÄ±!")
            print(f"   Kontrol edilen dizin: {os.path.abspath(data_dir)}")
            if os.path.exists(data_dir):
                try:
                    dir_contents = os.listdir(data_dir)
                    print(f"   Dizin iÃ§eriÄŸi ({len(dir_contents)} Ã¶ÄŸe):")
                    for item in dir_contents[:10]:
                        item_path = os.path.join(data_dir, item)
                        is_dir = os.path.isdir(item_path)
                        print(f"     - {item} {'[KLASÃ–R]' if is_dir else '[DOSYA]'}")
                except Exception as e:
                    print(f"   Dizin iÃ§eriÄŸi okunamadÄ±: {e}")
            print(f"\n   Aranan klasÃ¶rler:")
            for cls in APPLE_CLASSES:
                print(f"     - {cls}")
        else:
            print("\n--- SÄ±nÄ±f BazlÄ± Ã–rnek SayÄ±larÄ± ---")
            for class_name, class_idx in self.class_to_idx.items():
                count = sum(1 for _, label in self.samples if label == class_idx)
                print(f"   {class_name}: {count}")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Hata: {img_path} yÃ¼klenemedi: {e}")
            # BoÅŸ bir gÃ¶rÃ¼ntÃ¼ dÃ¶ndÃ¼r
            image = Image.new('RGB', (224, 224), color='black')

        if self.transform:
            image = self.transform(image)

        return image, label


class TransformSubset(Dataset):
    """random_split sonrasÄ±nda farklÄ± transform'lar uygulamak iÃ§in wrapper"""
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform

    def __getitem__(self, index):
        x, y = self.subset[index]  # AppleDataset'ten (PIL, label) dÃ¶ner
        if self.transform:
            x = self.transform(x)
        return x, y

    def __len__(self):
        return len(self.subset)


# --- 3. YARDIMCI FONKSIYONLAR ---

def create_split_table(dataset, tr_sub, val_sub, te_sub, class_names):
    stats = {cls: {'Train': 0, 'Val': 0, 'Test': 0} for cls in class_names}

    def fill(subset, col):
        for idx in subset.indices:
            _, label = dataset[idx]
            stats[class_names[label]][col] += 1

    fill(tr_sub, 'Train')
    fill(val_sub, 'Val')
    fill(te_sub, 'Test')

    df = pd.DataFrame.from_dict(stats, orient='index')
    df['Total'] = df.sum(axis=1)
    df = df[['Total', 'Train', 'Val', 'Test']]
    df.loc['GENEL TOPLAM'] = df.sum()
    return df


def unnormalize(tensor):
    tensor = tensor.cpu().clone()
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
    tensor = tensor * std + mean
    return tensor.clamp(0, 1)


def train_one_epoch(model, loader, criterion, optimizer, device, scaler=None, epoch_idx=None, phase_name="Phase1"):
    model.train()
    run_loss, correct, total = 0.0, 0, 0

    for batch_idx, (inputs, labels) in enumerate(loader):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        if scaler is not None:
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        run_loss += loss.item() * inputs.size(0)
        _, pred = torch.max(outputs, 1)
        correct += (pred == labels).sum().item()
        total += labels.size(0)

        if (batch_idx + 1) % 10 == 0:
            batch_acc = 100 * (pred == labels).sum().item() / labels.size(0)
            print(
                f"  â³ [{phase_name}] Epoch {epoch_idx} - Batch {batch_idx+1}/{len(loader)} "
                f"| Loss: {loss.item():.4f} | Acc: {batch_acc:.2f}%", end='\r'
            )

    ep_loss = run_loss / len(loader.dataset)
    ep_acc = 100 * correct / total
    return ep_loss, ep_acc


def eval_one_epoch(model, loader, criterion, device, scaler=None, phase_name="Val"):
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)

            if scaler is not None:
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
            else:
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            val_loss += loss.item() * inputs.size(0)
            _, pred = torch.max(outputs, 1)
            val_correct += (pred == labels).sum().item()
            val_total += labels.size(0)

    ep_val_loss = val_loss / len(loader.dataset)
    ep_val_acc = 100 * val_correct / val_total
    return ep_val_loss, ep_val_acc


def main():
    # --- CÄ°HAZ VE AMP AYARLARI ---
    if torch.cuda.is_available():
        device = torch.device("cuda")
        print(f"âœ… GPU: CUDA kullanÄ±lÄ±yor ({torch.cuda.get_device_name(0)})")
        print(f"   GPU Bellek: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
        torch.cuda.empty_cache()
    else:
        device = torch.device("cpu")
        print("âš ï¸  GPU bulunamadÄ±, CPU kullanÄ±lÄ±yor")
        print("   âš ï¸  Colab'da GPU'yu aktifleÅŸtirmeyi unutmayÄ±n!")

    print(f"\n--- SÄ°STEM AYARLARI ---")
    print(f"ğŸ–¥ï¸  Cihaz: {device}")
    print(f"ğŸ“š Mod: {'Transfer Learning (Pretrained ConvNeXt-Tiny)' if USE_TRANSFER_LEARNING else 'SÄ±fÄ±rdan EÄŸitim'}")
    print(f"ğŸ¯ SÄ±nÄ±f SayÄ±sÄ±: 4 (Multi-Class Classification)")
    print(f"ğŸ“¦ Batch Size: {BATCH_SIZE}")
    print(f"âš¡ Mixed Precision: {'Aktif (CUDA)' if USE_MIXED_PRECISION and device.type == 'cuda' else 'Pasif'}")
    print("-" * 50)

    scaler = torch.cuda.amp.GradScaler() if (USE_MIXED_PRECISION and device.type == "cuda") else None

    # --- VERÄ° DÄ°ZÄ°NÄ° KONTROL ---
    if not os.path.exists(DATA_DIR):
        print(f"âŒ HATA: Veri dizini bulunamadÄ±: {DATA_DIR}")
        print(f"   LÃ¼tfen Colab'da Google Drive'Ä± mount edin ve veri yolunu kontrol edin.")
        return

    print(f"\nğŸ“ Veri dizini: {DATA_DIR}")
    print(f"   Dizin mevcut: {os.path.exists(DATA_DIR)}")

    # --- VERÄ° SETÄ°NÄ° YÃœKLE ---
    try:
        raw_dataset = AppleDataset(DATA_DIR)
        classes = APPLE_CLASSES
        num_classes = len(classes)
    except Exception as e:
        print(f"âŒ HATA: Veri seti okunamadÄ±. {e}")
        import traceback
        traceback.print_exc()
        return

    total_len = len(raw_dataset)
    if total_len == 0:
        print("\nâŒ HATA: Veri seti boÅŸ! HiÃ§ resim bulunamadÄ±.")
        return
    if total_len < 10:
        print(f"\nâŒ HATA: Veri seti Ã§ok kÃ¼Ã§Ã¼k ({total_len} Ã¶rnek). En az 10 Ã¶rnek gerekli.")
        return

    # --- VERÄ°YÄ° TRAIN/VAL/TEST OLARAK BÃ–L ---
    train_len = int(total_len * 0.70)
    val_len = int(total_len * 0.15)
    test_len = total_len - train_len - val_len

    if train_len == 0 or val_len == 0 or test_len == 0:
        print(f"\nâŒ HATA: Veri bÃ¶lme hatasÄ± - Train: {train_len}, Val: {val_len}, Test: {test_len}")
        print(f"   Toplam Ã¶rnek sayÄ±sÄ± Ã§ok az: {total_len}")
        return

    try:
        train_subset, val_subset, test_subset = random_split(
            raw_dataset, [train_len, val_len, test_len],
            generator=torch.Generator().manual_seed(42)
        )
        print(f"\nâœ… Veri bÃ¶lme baÅŸarÄ±lÄ±:")
        print(f"   Train: {train_len}, Val: {val_len}, Test: {test_len}")
    except Exception as e:
        print(f"\nâŒ HATA: Veri bÃ¶lme iÅŸlemi baÅŸarÄ±sÄ±z: {e}")
        import traceback
        traceback.print_exc()
        return

    # --- VERÄ° SETÄ° DAÄILIM TABLOSU ---
    print("\n--- Veri Seti DaÄŸÄ±lÄ±m Tablosu ---")
    print(create_split_table(raw_dataset, train_subset, val_subset, test_subset, classes))
    print("-" * 50)

    # --- TRANSFORMLAR ---
    # ConvNeXt iÃ§in standart: RandomResizedCrop + Normalize
    train_transforms = transforms.Compose([
        transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(10),
        transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.1, hue=0.02),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    val_test_transforms = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    train_data = TransformSubset(train_subset, transform=train_transforms)
    val_data = TransformSubset(val_subset, transform=val_test_transforms)
    test_data = TransformSubset(test_subset, transform=val_test_transforms)

    # --- DATALOADER AYARLARI ---
    if torch.cuda.is_available():
        num_workers = 4
        pin_memory = True
    else:
        num_workers = 2
        pin_memory = False
    print(f"ğŸ“Š DataLoader AyarlarÄ±: num_workers={num_workers}, pin_memory={pin_memory}")

    train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True,
                              num_workers=num_workers, pin_memory=pin_memory)
    val_loader = DataLoader(val_data, batch_size=BATCH_SIZE, shuffle=False,
                            num_workers=num_workers, pin_memory=pin_memory)
    test_loader = DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=False,
                             num_workers=num_workers, pin_memory=pin_memory)

    # --- 4. MODEL MÄ°MARÄ°SÄ° (ConvNeXt-Tiny) ---
    if USE_TRANSFER_LEARNING:
        print("\n--- Model: Pretrained ConvNeXt-Tiny (Transfer Learning) ---")
        model = models.convnext_tiny(weights=ConvNeXt_Tiny_Weights.DEFAULT)

        # Feature extractor'Ä± dondur (Phase-1 iÃ§in)
        for param in model.features.parameters():
            param.requires_grad = False

        # Classifier'Ä± deÄŸiÅŸtir - 4 sÄ±nÄ±f iÃ§in
        num_ftrs = model.classifier[2].in_features
        model.classifier[2] = nn.Linear(num_ftrs, num_classes)
        print("âœ… Transfer Learning aktif - Phase-1: Sadece classifier eÄŸitilecek")
    else:
        print("\n--- Model: ConvNeXt-Tiny (SÄ±fÄ±rdan EÄŸitim) ---")
        model = models.convnext_tiny(weights=None)
        num_ftrs = model.classifier[2].in_features
        model.classifier[2] = nn.Linear(num_ftrs, num_classes)

    model = model.to(device)
    if device.type == "cuda":
        torch.cuda.empty_cache()

    print("\n--- DETAYLI MODEL MÄ°MARÄ°SÄ° ---")
    print(model)
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\nToplam Parametre SayÄ±sÄ±: {total_params:,}")
    print(f"EÄŸitilebilir Parametre SayÄ±sÄ±: {trainable_params:,}")
    print("-" * 40)

    # --- 5. LOSS FONKSÄ°YONU (Label Smoothing) ---
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

    # --- 6. EÄÄ°TÄ°M TARÄ°HÃ‡ESÄ° ---
    history = {
        'phase': [],
        'epoch': [],
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': []
    }

    # --- 7. PHASE-1: SADECE CLASSIFIER EÄÄ°TÄ°MÄ° ---
    if USE_TRANSFER_LEARNING:
        optimizer = optim.AdamW(model.classifier.parameters(),
                                lr=BASE_LR_PHASE1,
                                weight_decay=WEIGHT_DECAY)
        scheduler_phase1 = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.1, patience=5
        )
        total_epochs = NUM_EPOCHS_PHASE1 + (NUM_EPOCHS_PHASE2 if USE_TRANSFER_LEARNING else 0)

        print(f"\n{'='*70}")
        print(f"ğŸš€ PHASE-1 BAÅLIYOR: SADECE CLASSIFIER EÄÄ°TÄ°LECEK ({NUM_EPOCHS_PHASE1} EPOCH)")
        print(f"{'='*70}")
        start_time = time.time()

        for epoch in range(1, NUM_EPOCHS_PHASE1 + 1):
            epoch_start_time = time.time()
            print(f"\n{'â”€'*60}")
            print(f"ğŸ“Š [PHASE-1] EPOCH {epoch}/{total_epochs} - EÄÄ°TÄ°M BAÅLADI...")
            print(f"{'â”€'*60}")

            train_loss, train_acc = train_one_epoch(
                model, train_loader, criterion, optimizer, device,
                scaler=scaler, epoch_idx=epoch, phase_name="Phase-1"
            )

            if device.type == "cuda":
                torch.cuda.empty_cache()

            val_loss, val_acc = eval_one_epoch(
                model, val_loader, criterion, device,
                scaler=scaler, phase_name="Val-Phase1"
            )

            scheduler_phase1.step(val_loss)
            current_lr = optimizer.param_groups[0]['lr']

            epoch_time = time.time() - epoch_start_time

            # History kaydÄ±
            history['phase'].append('Phase-1')
            history['epoch'].append(epoch)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)
            history['val_loss'].append(val_loss)
            history['val_acc'].append(val_acc)

            print(f"\n  âœ… [PHASE-1] Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
            print(f"  ğŸ” [PHASE-1] Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")
            print(f"  ğŸ“š Learning Rate: {current_lr:.6f}")
            print(f"  â±ï¸  Epoch SÃ¼resi: {epoch_time:.2f} saniye")
            print(f"{'â”€'*60}")

        phase1_time = time.time() - start_time
        print(f"\nâ±ï¸  PHASE-1 Toplam SÃ¼re: {phase1_time/60:.2f} dakika ({phase1_time:.2f} saniye)")

    else:
        # SÄ±fÄ±rdan eÄŸitim senaryosu (tek faz)
        optimizer = optim.SGD(model.parameters(),
                              lr=BASE_LR_PHASE1,
                              momentum=0.9,
                              weight_decay=WEIGHT_DECAY)
        scheduler_phase1 = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.1, patience=5
        )
        total_epochs = NUM_EPOCHS_PHASE1

        print(f"\n{'='*70}")
        print(f"ğŸš€ EÄÄ°TÄ°M BAÅLIYOR (SIFIRDAN) - {NUM_EPOCHS_PHASE1} EPOCH")
        print(f"{'='*70}")
        start_time = time.time()

        for epoch in range(1, NUM_EPOCHS_PHASE1 + 1):
            epoch_start_time = time.time()
            print(f"\n{'â”€'*60}")
            print(f"ğŸ“Š [FULL] EPOCH {epoch}/{total_epochs} - EÄÄ°TÄ°M BAÅLADI...")
            print(f"{'â”€'*60}")

            train_loss, train_acc = train_one_epoch(
                model, train_loader, criterion, optimizer, device,
                scaler=scaler, epoch_idx=epoch, phase_name="Full"
            )

            if device.type == "cuda":
                torch.cuda.empty_cache()

            val_loss, val_acc = eval_one_epoch(
                model, val_loader, criterion, device,
                scaler=scaler, phase_name="Val-Full"
            )

            scheduler_phase1.step(val_loss)
            current_lr = optimizer.param_groups[0]['lr']

            epoch_time = time.time() - epoch_start_time

            history['phase'].append('Full')
            history['epoch'].append(epoch)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)
            history['val_loss'].append(val_loss)
            history['val_acc'].append(val_acc)

            print(f"\n  âœ… [FULL] Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
            print(f"  ğŸ” [FULL] Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")
            print(f"  ğŸ“š Learning Rate: {current_lr:.6f}")
            print(f"  â±ï¸  Epoch SÃ¼resi: {epoch_time:.2f} saniye")
            print(f"{'â”€'*60}")

        total_time = time.time() - start_time
        print(f"\nâ±ï¸  EÄÄ°TÄ°M Toplam SÃ¼re: {total_time/60:.2f} dakika ({total_time:.2f} saniye)")

    # --- 8. PHASE-2: FINE-TUNING (SON BLOKLAR) ---
    if USE_TRANSFER_LEARNING and NUM_EPOCHS_PHASE2 > 0:
        print(f"\n{'='*70}")
        print(f"ğŸ”§ PHASE-2 BAÅLIYOR: SON BLOKLAR FINE-TUNING ({NUM_EPOCHS_PHASE2} EPOCH)")
        print(f"{'='*70}")

        # Son 2 stage'i (features[3] ve features[4]) unfreeze et
        for param in model.features[3].parameters():
            param.requires_grad = True
        for param in model.features[4].parameters():
            param.requires_grad = True

        # Trainable parametreleri belirle
        trainable_params = [p for p in model.parameters() if p.requires_grad]
        trainable_count = sum(p.numel() for p in trainable_params)
        print(f"   Fine-Tuning iÃ§in EÄŸitilebilir Parametre SayÄ±sÄ±: {trainable_count:,}")

        # AdamW + CosineAnnealingLR
        optimizer_ft = optim.AdamW(trainable_params,
                                   lr=BASE_LR_PHASE2,
                                   weight_decay=WEIGHT_DECAY)
        scheduler_phase2 = optim.lr_scheduler.CosineAnnealingLR(
            optimizer_ft, T_max=NUM_EPOCHS_PHASE2
        )

        start_phase2 = time.time()
        for e in range(1, NUM_EPOCHS_PHASE2 + 1):
            global_epoch = NUM_EPOCHS_PHASE1 + e
            epoch_start_time = time.time()
            print(f"\n{'â”€'*60}")
            print(f"ğŸ“Š [PHASE-2] EPOCH {global_epoch}/{NUM_EPOCHS_PHASE1 + NUM_EPOCHS_PHASE2} - FINE-TUNING")
            print(f"{'â”€'*60}")

            train_loss, train_acc = train_one_epoch(
                model, train_loader, criterion, optimizer_ft, device,
                scaler=scaler, epoch_idx=global_epoch, phase_name="Phase-2"
            )

            if device.type == "cuda":
                torch.cuda.empty_cache()

            val_loss, val_acc = eval_one_epoch(
                model, val_loader, criterion, device,
                scaler=scaler, phase_name="Val-Phase2"
            )

            scheduler_phase2.step()
            current_lr = optimizer_ft.param_groups[0]['lr']
            epoch_time = time.time() - epoch_start_time

            history['phase'].append('Phase-2')
            history['epoch'].append(global_epoch)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)
            history['val_loss'].append(val_loss)
            history['val_acc'].append(val_acc)

            print(f"\n  âœ… [PHASE-2] Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
            print(f"  ğŸ” [PHASE-2] Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")
            print(f"  ğŸ“š Learning Rate: {current_lr:.6f}")
            print(f"  â±ï¸  Epoch SÃ¼resi: {epoch_time:.2f} saniye")
            print(f"{'â”€'*60}")

        phase2_time = time.time() - start_phase2
        print(f"\nâ±ï¸  PHASE-2 Toplam SÃ¼re: {phase2_time/60:.2f} dakika ({phase2_time:.2f} saniye)")

        total_time = (phase1_time if USE_TRANSFER_LEARNING else 0) + phase2_time
        print(f"\n{'='*70}")
        print(f"ğŸ‰ TÃœM EÄÄ°TÄ°M TAMAMLANDI!")
        print(f"   â±ï¸  Toplam SÃ¼re: {total_time/60:.2f} dakika ({total_time:.2f} saniye)")
        print(f"{'='*70}")

    # --- 9. LOSS / ACC GRAFÄ°KLERÄ° ---
    print("\nğŸ“ˆ Loss ve Accuracy grafikleri Ã§iziliyor...")

    epochs = history['epoch']
    train_loss = history['train_loss']
    val_loss = history['val_loss']
    train_acc = history['train_acc']
    val_acc = history['val_acc']

    plt.figure(figsize=(14, 5))

    # Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_loss, 'b-o', label='Train Loss', linewidth=2, markersize=4)
    plt.plot(epochs, val_loss, 'r-o', label='Val Loss', linewidth=2, markersize=4)
    plt.title('Loss GrafiÄŸi', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)

    # Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_acc, 'b-o', label='Train Accuracy', linewidth=2, markersize=4)
    plt.plot(epochs, val_acc, 'r-o', label='Val Accuracy', linewidth=2, markersize=4)
    plt.title('Accuracy GrafiÄŸi', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Accuracy (%)', fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('training_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

    # --- 10. TEST SETÄ° DEÄERLENDÄ°RMESÄ° ---
    print("\n--- Test Seti DeÄŸerlendirmesi ---")
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            if scaler is not None:
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
            else:
                outputs = model(inputs)

            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    print("\nSÄ±nÄ±flandÄ±rma Raporu:")
    print(classification_report(all_labels, all_preds, target_names=classes))

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix (4 Classes)', fontsize=14, fontweight='bold')
    plt.ylabel('GerÃ§ek', fontsize=12)
    plt.xlabel('Tahmin', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    # --- 11. Ã–RNEK TAHMÄ°NLER (Her sÄ±nÄ±ftan 3 Ã¶rnek) ---
    print("\nğŸ“¸ Test setinden Ã¶rnek tahminler hazÄ±rlanÄ±yor...")
    samples_by_class = {0: [], 1: [], 2: [], 3: []}

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs_gpu = inputs.to(device)

            if scaler is not None:
                with torch.cuda.amp.autocast():
                    outputs = model(inputs_gpu)
            else:
                outputs = model(inputs_gpu)

            _, preds = torch.max(outputs, 1)

            for i in range(len(labels)):
                lbl = labels[i].item()
                if len(samples_by_class[lbl]) < 3:
                    samples_by_class[lbl].append(
                        (inputs[i], lbl, preds[i].item())
                    )

            if all(len(samples_by_class[k]) >= 3 for k in range(4)):
                break

    all_samples = []
    for class_idx in range(4):
        all_samples.extend(samples_by_class[class_idx])

    plt.figure(figsize=(18, 12))
    for i, (img, true_label, pred_label) in enumerate(all_samples):
        ax = plt.subplot(4, 3, i + 1)
        img_unnorm = unnormalize(img).permute(1, 2, 0).numpy()
        plt.imshow(img_unnorm)

        true_lbl = classes[true_label]
        pred_lbl = classes[pred_label]
        col = 'green' if true_lbl == pred_lbl else 'red'

        true_short = true_lbl.replace('Apple___', '')
        pred_short = pred_lbl.replace('Apple___', '')

        plt.title(f"GerÃ§ek: {true_short}\nTahmin: {pred_short}",
                  color=col, fontsize=10, fontweight='bold')
        plt.axis('off')

    plt.suptitle("Ã–rnek Tahminler - Apple Diseases (Her SÄ±nÄ±ftan 3'er Tane)",
                 fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig('sample_predictions.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("\n--- EÄŸitim TamamlandÄ±! ---")
    print("Grafikler kaydedildi:")
    print("- training_curves.png")
    print("- confusion_matrix.png")
    print("- sample_predictions.png")


if __name__ == '__main__':
    main()
```

```
# -*- coding: utf-8 -*-
"""
ConvNeXt-Tiny ile PlantVillage Dataset'inde Elma SÄ±nÄ±flandÄ±rmasÄ±
4 SÄ±nÄ±flÄ± Multi-Class Classification (Apple Diseases)

GELÄ°ÅTÄ°RÄ°LMÄ°Å SÃœRÃœM v2
- Two-Phase Transfer Learning (Phase-1: Only Classifier, Phase-2: Fine-Tuning son bloklar)
- AdamW optimizer + Cosine Annealing LR (Phase-2)
- Label Smoothing ile CrossEntropyLoss
- RandomResizedCrop + geliÅŸmiÅŸ augmentations
- Mixed Precision (AMP) ile hÄ±zlÄ± eÄŸitim (CUDA varsa)
- Google Colab iÃ§in optimize (CUDA GPU desteÄŸi)
"""

import os
import time
import glob
from collections import defaultdict

# Colab ortamÄ± mÄ± kontrol et ve Drive mount
IN_COLAB = False
try:
    from google.colab import drive
    IN_COLAB = True
    drive.mount('/content/drive', force_remount=False)
    print("âœ… Google Drive baÄŸlandÄ±")
except Exception:
    print("âš ï¸  Google Colab ortamÄ± deÄŸil, Drive mount atlandÄ±")

# PyTorch import'larÄ± - Runtime restart kontrolÃ¼ ile
try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from torch.utils.data import DataLoader, random_split, Dataset
    from torchvision import transforms, models
    from torchvision.models import ConvNeXt_Tiny_Weights
    print("âœ… PyTorch ve Torchvision baÅŸarÄ±yla yÃ¼klendi")
except RuntimeError as e:
    if "TORCH_LIBRARY" in str(e):
        print("âŒ HATA: PyTorch Ã§akÄ±ÅŸmasÄ± tespit edildi!")
        if IN_COLAB:
            print("   Runtime otomatik olarak yeniden baÅŸlatÄ±lÄ±yor...")
            print("   LÃ¼tfen bu hÃ¼creyi tekrar Ã§alÄ±ÅŸtÄ±rÄ±n!")
            os.kill(os.getpid(), 9)  # Runtime'Ä± restart et
        else:
            print("   LÃ¼tfen Colab'da Runtime > Restart Runtime yapÄ±n ve tekrar Ã§alÄ±ÅŸtÄ±rÄ±n.")
            raise
    else:
        raise
except Exception as e:
    print(f"âŒ Import hatasÄ±: {e}")
    raise

from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# --- 1. AYARLAR ---

# Transfer Learning: True = HÄ±zlÄ± eÄŸitim (Pretrained), False = SÄ±fÄ±rdan
USE_TRANSFER_LEARNING = True

# PlantVillage dataset yolu - Colab iÃ§in
# Ã–rn: "/content/drive/MyDrive/PlantVillage/color"
DATA_DIR = "/content/drive/MyDrive/color"  # BURAYI KENDÄ° YOLUNA GÃ–RE GÃœNCELLE

# EÄŸitim parametreleri
NUM_EPOCHS_PHASE1 = 40        # Sadece classifier eÄŸitimi
NUM_EPOCHS_PHASE2 = 10        # Fine-tuning (son bloklar)
BATCH_SIZE = 32               # Colab CUDA iÃ§in optimize
BASE_LR_PHASE1 = 1e-4         # Phase-1 LR (sadece classifier)
BASE_LR_PHASE2 = 5e-6         # Phase-2 LR (fine-tuning)
WEIGHT_DECAY = 5e-4
USE_MIXED_PRECISION = True    # Colab CUDA iÃ§in hÄ±z ve bellek optimizasyonu

# Apple sÄ±nÄ±flarÄ±
APPLE_CLASSES = [
    'Apple___Apple_scab',
    'Apple___Black_rot',
    'Apple___Cedar_apple_rust',
    'Apple___healthy'
]


# --- 2. DATASET SINIFI ---

class AppleDataset(Dataset):
    """Elma verilerini 4 sÄ±nÄ±flÄ± multi-class olarak yÃ¼kleyen custom dataset"""
    def __init__(self, data_dir, transform=None):
        self.data_dir = data_dir
        self.transform = transform
        self.samples = []
        self.class_to_idx = {
            'Apple___Apple_scab': 0,
            'Apple___Black_rot': 1,
            'Apple___Cedar_apple_rust': 2,
            'Apple___healthy': 3
        }
        self.idx_to_class = {v: k for k, v in self.class_to_idx.items()}

        found_any = False
        print("\n--- Veri KlasÃ¶rleri TaranÄ±yor ---")
        for folder_name in APPLE_CLASSES:
            folder_path = os.path.join(data_dir, folder_name)
            if os.path.exists(folder_path):
                # FarklÄ± uzantÄ±larÄ± dene
                images = (
                    glob.glob(os.path.join(folder_path, '*.JPG')) +
                    glob.glob(os.path.join(folder_path, '*.jpg')) +
                    glob.glob(os.path.join(folder_path, '*.jpeg')) +
                    glob.glob(os.path.join(folder_path, '*.png'))
                )

                if len(images) == 0:
                    try:
                        all_files = os.listdir(folder_path)
                        print(f"âš ï¸  {folder_name}: KlasÃ¶r var ama resim bulunamadÄ±")
                        print(f"   KlasÃ¶rdeki dosya sayÄ±sÄ±: {len(all_files)}")
                        if len(all_files) > 0:
                            print(f"   Ä°lk 5 dosya: {all_files[:5]}")
                    except Exception:
                        pass
                else:
                    label = self.class_to_idx[folder_name]
                    image_count = 0
                    for img_path in images:
                        self.samples.append((img_path, label))
                        image_count += 1
                    print(f"âœ… {folder_name}: {image_count} resim bulundu")
                    found_any = True
            else:
                print(f"âŒ KlasÃ¶r bulunamadÄ±: {folder_path}")

        print(f"\nğŸ“Š Toplam Ã¶rnek sayÄ±sÄ±: {len(self.samples)}")
        if len(self.samples) == 0:
            print("\nâŒ HATA: HiÃ§ resim bulunamadÄ±!")
            print(f"   Kontrol edilen dizin: {os.path.abspath(data_dir)}")
            if os.path.exists(data_dir):
                try:
                    dir_contents = os.listdir(data_dir)
                    print(f"   Dizin iÃ§eriÄŸi ({len(dir_contents)} Ã¶ÄŸe):")
                    for item in dir_contents[:10]:
                        item_path = os.path.join(data_dir, item)
                        is_dir = os.path.isdir(item_path)
                        print(f"     - {item} {'[KLASÃ–R]' if is_dir else '[DOSYA]'}")
                except Exception as e:
                    print(f"   Dizin iÃ§eriÄŸi okunamadÄ±: {e}")
            print(f"\n   Aranan klasÃ¶rler:")
            for cls in APPLE_CLASSES:
                print(f"     - {cls}")
        else:
            print("\n--- SÄ±nÄ±f BazlÄ± Ã–rnek SayÄ±larÄ± ---")
            for class_name, class_idx in self.class_to_idx.items():
                count = sum(1 for _, label in self.samples if label == class_idx)
                print(f"   {class_name}: {count}")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Hata: {img_path} yÃ¼klenemedi: {e}")
            # BoÅŸ bir gÃ¶rÃ¼ntÃ¼ dÃ¶ndÃ¼r
            image = Image.new('RGB', (224, 224), color='black')

        if self.transform:
            image = self.transform(image)

        return image, label


class TransformSubset(Dataset):
    """random_split sonrasÄ±nda farklÄ± transform'lar uygulamak iÃ§in wrapper"""
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform

    def __getitem__(self, index):
        x, y = self.subset[index]  # AppleDataset'ten (PIL, label) dÃ¶ner
        if self.transform:
            x = self.transform(x)
        return x, y

    def __len__(self):
        return len(self.subset)


# --- 3. YARDIMCI FONKSIYONLAR ---

def create_split_table(dataset, tr_sub, val_sub, te_sub, class_names):
    stats = {cls: {'Train': 0, 'Val': 0, 'Test': 0} for cls in class_names}

    def fill(subset, col):
        for idx in subset.indices:
            _, label = dataset[idx]
            stats[class_names[label]][col] += 1

    fill(tr_sub, 'Train')
    fill(val_sub, 'Val')
    fill(te_sub, 'Test')

    df = pd.DataFrame.from_dict(stats, orient='index')
    df['Total'] = df.sum(axis=1)
    df = df[['Total', 'Train', 'Val', 'Test']]
    df.loc['GENEL TOPLAM'] = df.sum()
    return df


def unnormalize(tensor):
    tensor = tensor.cpu().clone()
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
    tensor = tensor * std + mean
    return tensor.clamp(0, 1)


def train_one_epoch(model, loader, criterion, optimizer, device, scaler=None, epoch_idx=None, phase_name="Phase1"):
    model.train()
    run_loss, correct, total = 0.0, 0, 0

    for batch_idx, (inputs, labels) in enumerate(loader):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        if scaler is not None:
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        run_loss += loss.item() * inputs.size(0)
        _, pred = torch.max(outputs, 1)
        correct += (pred == labels).sum().item()
        total += labels.size(0)

        if (batch_idx + 1) % 10 == 0:
            batch_acc = 100 * (pred == labels).sum().item() / labels.size(0)
            print(
                f"  â³ [{phase_name}] Epoch {epoch_idx} - Batch {batch_idx+1}/{len(loader)} "
                f"| Loss: {loss.item():.4f} | Acc: {batch_acc:.2f}%", end='\r'
            )

    ep_loss = run_loss / len(loader.dataset)
    ep_acc = 100 * correct / total
    return ep_loss, ep_acc


def eval_one_epoch(model, loader, criterion, device, scaler=None, phase_name="Val"):
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)

            if scaler is not None:
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
            else:
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            val_loss += loss.item() * inputs.size(0)
            _, pred = torch.max(outputs, 1)
            val_correct += (pred == labels).sum().item()
            val_total += labels.size(0)

    ep_val_loss = val_loss / len(loader.dataset)
    ep_val_acc = 100 * val_correct / val_total
    return ep_val_loss, ep_val_acc


def main():
    # --- CÄ°HAZ VE AMP AYARLARI ---
    if torch.cuda.is_available():
        device = torch.device("cuda")
        print(f"âœ… GPU: CUDA kullanÄ±lÄ±yor ({torch.cuda.get_device_name(0)})")
        print(f"   GPU Bellek: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
        torch.cuda.empty_cache()
    else:
        device = torch.device("cpu")
        print("âš ï¸  GPU bulunamadÄ±, CPU kullanÄ±lÄ±yor")
        print("   âš ï¸  Colab'da GPU'yu aktifleÅŸtirmeyi unutmayÄ±n!")

    print(f"\n--- SÄ°STEM AYARLARI ---")
    print(f"ğŸ–¥ï¸  Cihaz: {device}")
    print(f"ğŸ“š Mod: {'Transfer Learning (Pretrained ConvNeXt-Tiny)' if USE_TRANSFER_LEARNING else 'SÄ±fÄ±rdan EÄŸitim'}")
    print(f"ğŸ¯ SÄ±nÄ±f SayÄ±sÄ±: 4 (Multi-Class Classification)")
    print(f"ğŸ“¦ Batch Size: {BATCH_SIZE}")
    print(f"âš¡ Mixed Precision: {'Aktif (CUDA)' if USE_MIXED_PRECISION and device.type == 'cuda' else 'Pasif'}")
    print("-" * 50)

    scaler = torch.cuda.amp.GradScaler() if (USE_MIXED_PRECISION and device.type == "cuda") else None

    # --- VERÄ° DÄ°ZÄ°NÄ° KONTROL ---
    if not os.path.exists(DATA_DIR):
        print(f"âŒ HATA: Veri dizini bulunamadÄ±: {DATA_DIR}")
        print(f"   LÃ¼tfen Colab'da Google Drive'Ä± mount edin ve veri yolunu kontrol edin.")
        return

    print(f"\nğŸ“ Veri dizini: {DATA_DIR}")
    print(f"   Dizin mevcut: {os.path.exists(DATA_DIR)}")

    # --- VERÄ° SETÄ°NÄ° YÃœKLE ---
    try:
        raw_dataset = AppleDataset(DATA_DIR)
        classes = APPLE_CLASSES
        num_classes = len(classes)
    except Exception as e:
        print(f"âŒ HATA: Veri seti okunamadÄ±. {e}")
        import traceback
        traceback.print_exc()
        return

    total_len = len(raw_dataset)
    if total_len == 0:
        print("\nâŒ HATA: Veri seti boÅŸ! HiÃ§ resim bulunamadÄ±.")
        return
    if total_len < 10:
        print(f"\nâŒ HATA: Veri seti Ã§ok kÃ¼Ã§Ã¼k ({total_len} Ã¶rnek). En az 10 Ã¶rnek gerekli.")
        return

    # --- VERÄ°YÄ° TRAIN/VAL/TEST OLARAK BÃ–L ---
    train_len = int(total_len * 0.70)
    val_len = int(total_len * 0.15)
    test_len = total_len - train_len - val_len

    if train_len == 0 or val_len == 0 or test_len == 0:
        print(f"\nâŒ HATA: Veri bÃ¶lme hatasÄ± - Train: {train_len}, Val: {val_len}, Test: {test_len}")
        print(f"   Toplam Ã¶rnek sayÄ±sÄ± Ã§ok az: {total_len}")
        return

    try:
        train_subset, val_subset, test_subset = random_split(
            raw_dataset, [train_len, val_len, test_len],
            generator=torch.Generator().manual_seed(42)
        )
        print(f"\nâœ… Veri bÃ¶lme baÅŸarÄ±lÄ±:")
        print(f"   Train: {train_len}, Val: {val_len}, Test: {test_len}")
    except Exception as e:
        print(f"\nâŒ HATA: Veri bÃ¶lme iÅŸlemi baÅŸarÄ±sÄ±z: {e}")
        import traceback
        traceback.print_exc()
        return

    # --- VERÄ° SETÄ° DAÄILIM TABLOSU ---
    print("\n--- Veri Seti DaÄŸÄ±lÄ±m Tablosu ---")
    print(create_split_table(raw_dataset, train_subset, val_subset, test_subset, classes))
    print("-" * 50)

    # --- TRANSFORMLAR ---
    # ConvNeXt iÃ§in standart: RandomResizedCrop + Normalize
    train_transforms = transforms.Compose([
        transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(10),
        transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.1, hue=0.02),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    val_test_transforms = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    train_data = TransformSubset(train_subset, transform=train_transforms)
    val_data = TransformSubset(val_subset, transform=val_test_transforms)
    test_data = TransformSubset(test_subset, transform=val_test_transforms)

    # --- DATALOADER AYARLARI ---
    if torch.cuda.is_available():
        num_workers = 4
        pin_memory = True
    else:
        num_workers = 2
        pin_memory = False
    print(f"ğŸ“Š DataLoader AyarlarÄ±: num_workers={num_workers}, pin_memory={pin_memory}")

    train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True,
                              num_workers=num_workers, pin_memory=pin_memory)
    val_loader = DataLoader(val_data, batch_size=BATCH_SIZE, shuffle=False,
                            num_workers=num_workers, pin_memory=pin_memory)
    test_loader = DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=False,
                             num_workers=num_workers, pin_memory=pin_memory)

    # --- 4. MODEL MÄ°MARÄ°SÄ° (ConvNeXt-Tiny) ---
    if USE_TRANSFER_LEARNING:
        print("\n--- Model: Pretrained ConvNeXt-Tiny (Transfer Learning) ---")
        model = models.convnext_tiny(weights=ConvNeXt_Tiny_Weights.DEFAULT)

        # Feature extractor'Ä± dondur (Phase-1 iÃ§in)
        for param in model.features.parameters():
            param.requires_grad = False

        # Classifier'Ä± deÄŸiÅŸtir - 4 sÄ±nÄ±f iÃ§in
        num_ftrs = model.classifier[2].in_features
        model.classifier[2] = nn.Linear(num_ftrs, num_classes)
        print("âœ… Transfer Learning aktif - Phase-1: Sadece classifier eÄŸitilecek")
    else:
        print("\n--- Model: ConvNeXt-Tiny (SÄ±fÄ±rdan EÄŸitim) ---")
        model = models.convnext_tiny(weights=None)
        num_ftrs = model.classifier[2].in_features
        model.classifier[2] = nn.Linear(num_ftrs, num_classes)

    model = model.to(device)
    if device.type == "cuda":
        torch.cuda.empty_cache()

    print("\n--- DETAYLI MODEL MÄ°MARÄ°SÄ° ---")
    print(model)
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\nToplam Parametre SayÄ±sÄ±: {total_params:,}")
    print(f"EÄŸitilebilir Parametre SayÄ±sÄ±: {trainable_params:,}")
    print("-" * 40)

    # --- 5. LOSS FONKSÄ°YONU (Label Smoothing) ---
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

    # --- 6. EÄÄ°TÄ°M TARÄ°HÃ‡ESÄ° ---
    history = {
        'phase': [],
        'epoch': [],
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': []
    }

    # --- 7. PHASE-1: SADECE CLASSIFIER EÄÄ°TÄ°MÄ° ---
    if USE_TRANSFER_LEARNING:
        optimizer = optim.AdamW(model.classifier.parameters(),
                                lr=BASE_LR_PHASE1,
                                weight_decay=WEIGHT_DECAY)
        scheduler_phase1 = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.1, patience=5
        )
        total_epochs = NUM_EPOCHS_PHASE1 + (NUM_EPOCHS_PHASE2 if USE_TRANSFER_LEARNING else 0)

        print(f"\n{'='*70}")
        print(f"ğŸš€ PHASE-1 BAÅLIYOR: SADECE CLASSIFIER EÄÄ°TÄ°LECEK ({NUM_EPOCHS_PHASE1} EPOCH)")
        print(f"{'='*70}")
        start_time = time.time()

        for epoch in range(1, NUM_EPOCHS_PHASE1 + 1):
            epoch_start_time = time.time()
            print(f"\n{'â”€'*60}")
            print(f"ğŸ“Š [PHASE-1] EPOCH {epoch}/{total_epochs} - EÄÄ°TÄ°M BAÅLADI...")
            print(f"{'â”€'*60}")

            train_loss, train_acc = train_one_epoch(
                model, train_loader, criterion, optimizer, device,
                scaler=scaler, epoch_idx=epoch, phase_name="Phase-1"
            )

            if device.type == "cuda":
                torch.cuda.empty_cache()

            val_loss, val_acc = eval_one_epoch(
                model, val_loader, criterion, device,
                scaler=scaler, phase_name="Val-Phase1"
            )

            scheduler_phase1.step(val_loss)
            current_lr = optimizer.param_groups[0]['lr']

            epoch_time = time.time() - epoch_start_time

            # History kaydÄ±
            history['phase'].append('Phase-1')
            history['epoch'].append(epoch)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)
            history['val_loss'].append(val_loss)
            history['val_acc'].append(val_acc)

            print(f"\n  âœ… [PHASE-1] Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
            print(f"  ğŸ” [PHASE-1] Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")
            print(f"  ğŸ“š Learning Rate: {current_lr:.6f}")
            print(f"  â±ï¸  Epoch SÃ¼resi: {epoch_time:.2f} saniye")
            print(f"{'â”€'*60}")

        phase1_time = time.time() - start_time
        print(f"\nâ±ï¸  PHASE-1 Toplam SÃ¼re: {phase1_time/60:.2f} dakika ({phase1_time:.2f} saniye)")

    else:
        # SÄ±fÄ±rdan eÄŸitim senaryosu (tek faz)
        optimizer = optim.SGD(model.parameters(),
                              lr=BASE_LR_PHASE1,
                              momentum=0.9,
                              weight_decay=WEIGHT_DECAY)
        scheduler_phase1 = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.1, patience=5
        )
        total_epochs = NUM_EPOCHS_PHASE1

        print(f"\n{'='*70}")
        print(f"ğŸš€ EÄÄ°TÄ°M BAÅLIYOR (SIFIRDAN) - {NUM_EPOCHS_PHASE1} EPOCH")
        print(f"{'='*70}")
        start_time = time.time()

        for epoch in range(1, NUM_EPOCHS_PHASE1 + 1):
            epoch_start_time = time.time()
            print(f"\n{'â”€'*60}")
            print(f"ğŸ“Š [FULL] EPOCH {epoch}/{total_epochs} - EÄÄ°TÄ°M BAÅLADI...")
            print(f"{'â”€'*60}")

            train_loss, train_acc = train_one_epoch(
                model, train_loader, criterion, optimizer, device,
                scaler=scaler, epoch_idx=epoch, phase_name="Full"
            )

            if device.type == "cuda":
                torch.cuda.empty_cache()

            val_loss, val_acc = eval_one_epoch(
                model, val_loader, criterion, device,
                scaler=scaler, phase_name="Val-Full"
            )

            scheduler_phase1.step(val_loss)
            current_lr = optimizer.param_groups[0]['lr']

            epoch_time = time.time() - epoch_start_time

            history['phase'].append('Full')
            history['epoch'].append(epoch)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)
            history['val_loss'].append(val_loss)
            history['val_acc'].append(val_acc)

            print(f"\n  âœ… [FULL] Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
            print(f"  ğŸ” [FULL] Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")
            print(f"  ğŸ“š Learning Rate: {current_lr:.6f}")
            print(f"  â±ï¸  Epoch SÃ¼resi: {epoch_time:.2f} saniye")
            print(f"{'â”€'*60}")

        total_time = time.time() - start_time
        print(f"\nâ±ï¸  EÄÄ°TÄ°M Toplam SÃ¼re: {total_time/60:.2f} dakika ({total_time:.2f} saniye)")

    # --- 8. PHASE-2: FINE-TUNING (SON BLOKLAR) ---
    if USE_TRANSFER_LEARNING and NUM_EPOCHS_PHASE2 > 0:
        print(f"\n{'='*70}")
        print(f"ğŸ”§ PHASE-2 BAÅLIYOR: SON BLOKLAR FINE-TUNING ({NUM_EPOCHS_PHASE2} EPOCH)")
        print(f"{'='*70}")

        # Son 2 stage'i (features[3] ve features[4]) unfreeze et
        for param in model.features[3].parameters():
            param.requires_grad = True
        for param in model.features[4].parameters():
            param.requires_grad = True

        # Trainable parametreleri belirle
        trainable_params = [p for p in model.parameters() if p.requires_grad]
        trainable_count = sum(p.numel() for p in trainable_params)
        print(f"   Fine-Tuning iÃ§in EÄŸitilebilir Parametre SayÄ±sÄ±: {trainable_count:,}")

        # AdamW + CosineAnnealingLR
        optimizer_ft = optim.AdamW(trainable_params,
                                   lr=BASE_LR_PHASE2,
                                   weight_decay=WEIGHT_DECAY)
        scheduler_phase2 = optim.lr_scheduler.CosineAnnealingLR(
            optimizer_ft, T_max=NUM_EPOCHS_PHASE2
        )

        start_phase2 = time.time()
        for e in range(1, NUM_EPOCHS_PHASE2 + 1):
            global_epoch = NUM_EPOCHS_PHASE1 + e
            epoch_start_time = time.time()
            print(f"\n{'â”€'*60}")
            print(f"ğŸ“Š [PHASE-2] EPOCH {global_epoch}/{NUM_EPOCHS_PHASE1 + NUM_EPOCHS_PHASE2} - FINE-TUNING")
            print(f"{'â”€'*60}")

            train_loss, train_acc = train_one_epoch(
                model, train_loader, criterion, optimizer_ft, device,
                scaler=scaler, epoch_idx=global_epoch, phase_name="Phase-2"
            )

            if device.type == "cuda":
                torch.cuda.empty_cache()

            val_loss, val_acc = eval_one_epoch(
                model, val_loader, criterion, device,
                scaler=scaler, phase_name="Val-Phase2"
            )

            scheduler_phase2.step()
            current_lr = optimizer_ft.param_groups[0]['lr']
            epoch_time = time.time() - epoch_start_time

            history['phase'].append('Phase-2')
            history['epoch'].append(global_epoch)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)
            history['val_loss'].append(val_loss)
            history['val_acc'].append(val_acc)

            print(f"\n  âœ… [PHASE-2] Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
            print(f"  ğŸ” [PHASE-2] Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")
            print(f"  ğŸ“š Learning Rate: {current_lr:.6f}")
            print(f"  â±ï¸  Epoch SÃ¼resi: {epoch_time:.2f} saniye")
            print(f"{'â”€'*60}")

        phase2_time = time.time() - start_phase2
        print(f"\nâ±ï¸  PHASE-2 Toplam SÃ¼re: {phase2_time/60:.2f} dakika ({phase2_time:.2f} saniye)")

        total_time = (phase1_time if USE_TRANSFER_LEARNING else 0) + phase2_time
        print(f"\n{'='*70}")
        print(f"ğŸ‰ TÃœM EÄÄ°TÄ°M TAMAMLANDI!")
        print(f"   â±ï¸  Toplam SÃ¼re: {total_time/60:.2f} dakika ({total_time:.2f} saniye)")
        print(f"{'='*70}")

    # --- 9. LOSS / ACC GRAFÄ°KLERÄ° ---
    print("\nğŸ“ˆ Loss ve Accuracy grafikleri Ã§iziliyor...")

    epochs = history['epoch']
    train_loss = history['train_loss']
    val_loss = history['val_loss']
    train_acc = history['train_acc']
    val_acc = history['val_acc']

    plt.figure(figsize=(14, 5))

    # Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_loss, 'b-o', label='Train Loss', linewidth=2, markersize=4)
    plt.plot(epochs, val_loss, 'r-o', label='Val Loss', linewidth=2, markersize=4)
    plt.title('Loss GrafiÄŸi', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)

    # Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_acc, 'b-o', label='Train Accuracy', linewidth=2, markersize=4)
    plt.plot(epochs, val_acc, 'r-o', label='Val Accuracy', linewidth=2, markersize=4)
    plt.title('Accuracy GrafiÄŸi', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Accuracy (%)', fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('training_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

    # --- 10. TEST SETÄ° DEÄERLENDÄ°RMESÄ° ---
    print("\n--- Test Seti DeÄŸerlendirmesi ---")
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            if scaler is not None:
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
            else:
                outputs = model(inputs)

            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    print("\nSÄ±nÄ±flandÄ±rma Raporu:")
    print(classification_report(all_labels, all_preds, target_names=classes))

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix (4 Classes)', fontsize=14, fontweight='bold')
    plt.ylabel('GerÃ§ek', fontsize=12)
    plt.xlabel('Tahmin', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    # --- 11. Ã–RNEK TAHMÄ°NLER (Her sÄ±nÄ±ftan 3 Ã¶rnek) ---
    print("\nğŸ“¸ Test setinden Ã¶rnek tahminler hazÄ±rlanÄ±yor...")
    samples_by_class = {0: [], 1: [], 2: [], 3: []}

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs_gpu = inputs.to(device)

            if scaler is not None:
                with torch.cuda.amp.autocast():
                    outputs = model(inputs_gpu)
            else:
                outputs = model(inputs_gpu)

            _, preds = torch.max(outputs, 1)

            for i in range(len(labels)):
                lbl = labels[i].item()
                if len(samples_by_class[lbl]) < 3:
                    samples_by_class[lbl].append(
                        (inputs[i], lbl, preds[i].item())
                    )

            if all(len(samples_by_class[k]) >= 3 for k in range(4)):
                break

    all_samples = []
    for class_idx in range(4):
        all_samples.extend(samples_by_class[class_idx])

    plt.figure(figsize=(18, 12))
    for i, (img, true_label, pred_label) in enumerate(all_samples):
        ax = plt.subplot(4, 3, i + 1)
        img_unnorm = unnormalize(img).permute(1, 2, 0).numpy()
        plt.imshow(img_unnorm)

        true_lbl = classes[true_label]
        pred_lbl = classes[pred_label]
        col = 'green' if true_lbl == pred_lbl else 'red'

        true_short = true_lbl.replace('Apple___', '')
        pred_short = pred_lbl.replace('Apple___', '')

        plt.title(f"GerÃ§ek: {true_short}\nTahmin: {pred_short}",
                  color=col, fontsize=10, fontweight='bold')
        plt.axis('off')

    plt.suptitle("Ã–rnek Tahminler - Apple Diseases (Her SÄ±nÄ±ftan 3'er Tane)",
                 fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig('sample_predictions.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("\n--- EÄŸitim TamamlandÄ±! ---")
    print("Grafikler kaydedildi:")
    print("- training_curves.png")
    print("- confusion_matrix.png")
    print("- sample_predictions.png")


if __name__ == '__main__':
    main()
```
